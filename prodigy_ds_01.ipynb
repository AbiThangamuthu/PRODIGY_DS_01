```python
import pandas as pd
import matplotlib.pyplot as plt

# Example 1: Bar chart for the distribution of a categorical variable (e.g., 'Branch')
branch_counts = df['Branch'].value_counts()

plt.figure(figsize=(10, 6))
branch_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Branches')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# Example 2: Histogram for the distribution of a continuous variable (e.g., 'Unit price')
plt.figure(figsize=(10, 6))
plt.hist(df['Unit price'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
